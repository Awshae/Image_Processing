```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("Input Images/girl.jpeg", cv2.IMREAD_GRAYSCALE)

dft = np.fft.fft2(image)

N = image.shape[0]
M = image.shape[1]

cosw = np.zeros((N,N))

for i in range(N):
    for j in range(N):
        cosw[i,j] = np.cos((np.pi*(2*i+1)*j)/(2*N))

dct = np.matmul(np.matmul(cosw,image),cosw.T)

plt.subplot(221), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.axis('off')

plt.subplot(222), plt.imshow(np.log(1 + np.abs(dft)), cmap = 'gray')
plt.title('2-D DFT'), plt.axis('off')

plt.subplot(223), plt.imshow(dct, cmap = 'gray')
plt.title('2-D DCT'), plt.axis('off')

plt.show()
        
```
