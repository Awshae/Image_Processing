```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("Input Images/girl.jpeg", cv2.IMREAD_GRAYSCALE)

sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 5)
sobel_mag =  np.sqrt(sobelx**2 + sobely**2)

scharrx = cv2.Scharr(image, cv2.CV_64F, 1, 0)
scharry = cv2.Scharr(image, cv2.CV_64F, 0, 1)
scharr_mag =  np.sqrt(scharrx**2 + scharry**2)

log = cv2.GaussianBlur(image,(5,5),0)
log = cv2.Laplacian(log, cv2.CV_64F)

canny = cv2.Canny(image,100,200)

plt.subplot(231),plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.axis('off')

plt.subplot(232),plt.imshow(sobel_mag, cmap='gray')
plt.title('Noise Image'), plt.axis('off')

plt.subplot(233),plt.imshow(scharr_mag, cmap='gray')
plt.title('Mean Filtered'), plt.axis('off')

plt.subplot(234),plt.imshow(log, cmap='gray')
plt.title('Median Filtered'), plt.axis('off')

plt.subplot(235),plt.imshow(canny, cmap='gray')
plt.title('Gaussian Filtered'), plt.axis('off')

plt.show()
```
