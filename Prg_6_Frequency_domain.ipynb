```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fig = plt.figure(figsize=(10,7))
img= cv2.imread(r"Input Images/noise.jpeg",cv2.IMREAD_UNCHANGED)

rows= 2
columns = 2

fig = plt.figure(figsize=(10,7))
fig.add_subplot(rows,columns,1)

fname=r"Input Images/girl.jpeg"
image1 = Image.open(fname)

plt.imshow(image1)
fig.add_subplot(rows,columns,2)
kernel = np.ones((10,10),np.float32)/25
meanFilter = cv2.filter2D(img,-1,kernel)
plt.imshow(meanFilter)
plt.axis('on')

```

```python
#Low Pass
lowPass = cv2.filter2D(img, -1, kernel)
lowPass = img - lowPass
plt.imshow(lowPass)
```

```python
gaussBlur = cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT)
plt.imshow(gaussBlur)
```

```python
highPass = img - gaussBlur
plt.imshow(highPass)
```

```python
bandreject_image = lowPass + highPass
bandpass_image = img - bandreject_image
bandpas_image = np.absolute(bandpass_image)
plt.imshow(bandpass_image)
```
